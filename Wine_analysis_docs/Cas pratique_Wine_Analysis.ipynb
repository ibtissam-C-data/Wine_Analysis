```python
# Situation post Etape 1:
round(df_Wine.notna().mean()*100,2)
```

```python
# @title Colonne désignation :

# elle est remplie a hauteur de 71% et d'aprés ma lecture du df je constate que la colonne titre peut nous aider à la remplir.
# dans "title", la designation est positionnée après la date et avant "()"

import re

def extraction_designation(nom_complet):
    if pd.isna(nom_complet):
        return None

    resultat = re.search(r'\d{4}\s+([^(]+)', nom_complet)
    if resultat:
        return resultat.group(1)
    return None

# J'applique la fonction à la colonne designation
df_Wine['designation'] = df_Wine['title'].apply(extraction_designation)
```

## C- Création de nouvelles colonnes qui peuvent être interessantes dans notre analyse:

```python
# @title Colonne Continent :

# ceci me permet de visulaiser de façon macro l'approvisionnement du marché américain de vin par continent:

Dico_pays = {
"Amerique_Nord" : ['US', 'Mexico', 'Canada'],
"Amerique_Sud" : ['Argentina', 'Chile', 'Brazil', 'Uruguay', 'Peru'],
"Afrique" :['South Africa', 'Morocco', 'Egypt'],
"Europe" : ['Italy', 'Portugal', 'Spain', 'France', 'Germany', 'Austria', 'Hungary', 'Greece', 'Romania',
               'Czech Republic', 'Slovenia', 'Luxembourg', 'Croatia', 'England', 'Serbia', 'Moldova',
               'Bulgaria', 'Switzerland', 'Bosnia and Herzegovina', 'Ukraine', 'Slovakia', 'Macedonia'],
"Asie" : ['Israel', 'Turkey', 'Georgia', 'Lebanon', 'India', 'Cyprus', 'Armenia', 'China'],
'Oceanique': ['Australia', 'New Zealand']
}

def continent(country):
  for continent, countries in Dico_pays.items():
    if country in countries:
      return continent

    else:    "Inconnu"          # le else est mis par sécurité mais j'ai fait un copier/coller des pays du df.

df_Wine['continent'] = df_Wine['country'].apply(continent)
```

```python
df_Wine['continent'].unique()
```

```python
# @title Colonne Année : pour analyser le Millésime


def Extraction_Annee_V3(title):
  motif = re.compile(r"\b[12][0-9]{3}\b") # ici je cherche les année entre 1000 et 2999 et ne capter que les nombre à 4chiffres
  recherche = motif.findall(title)
  if 'NV' in title.strip():
    return 'NV'
  if recherche:
      if len(recherche) == 1:
          return int(recherche[0]) if int(recherche[0]) < 2025 else None
      else:
          # Je trie les années extraites
          recherche_sorted = sorted(map(int, recherche))
          # Je capte la valeur précédente de la valeur maximale si cette dernière est inf ou égale à 2025
          if recherche_sorted[-1] < 2025:
              return recherche_sorted[-1]
          else:
              previous_annee = recherche_sorted[-2] if len(recherche_sorted) > 1 else None
              return previous_annee if previous_annee < 2025 else None
  return None


df_Wine['Année'] = df_Wine['title'].apply(Extraction_Annee_V3)

```

```python
df_Wine[df_Wine['Année'] == 'NV']['Année'].value_counts()
```

```python
# 3841 vin en "NV" , il s'agit de bouteilles de vin effervescent (champagne, mousseau ..etc)

# du fait que je ne maitrise pas le monde des vins, j'ai fait des recherche pour comprendre la notation NV
# NV => "Non-Vintage" en anglais et "sans millésime" en français.
# Il s'agit d'un assemblage de raisins provenant de plusieurs années de production.
# Seuls les vins effervescents sont concernés par cette classification NV.
# D'ailleur nous pouvons remarquer que dans les titre nous avons l'indication "Champagne" pour les français.
```

```python
df_Wine[df_Wine['Année'] == 'NV'].head(3)
```

```python
# @title Suite traitement année => Todatetime


# Je souhaite exploiter l'info NV => je créé une colonne avec un Bool:
df_Wine['Vin_effervescant'] = df_Wine['Année'] == 'NV'

# Je remplace les NV en Not (equivalent du Nan pour pour les date Time):
df_Wine['Année'] = df_Wine['Année'].replace('NV', np.nan)
#je converti en datetime:
df_Wine['Année'] = pd.to_datetime(df_Wine['Année'], format='%Y')
```
